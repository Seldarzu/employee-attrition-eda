Aykırı değer (outlier) tespiti  
Bu adımda sayısal değişkenlerde **IQR yöntemi** ile uç değerleri inceleyeceğiz.


```python
# 🔹 Adım 3 — Aykırı Değer Analizi (IQR yöntemi)

outlier_report = []

# Sadece sayısal sütunlarda kontrol
num_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()

for col in num_cols:
    series = df[col].dropna()
    if series.empty:
        continue
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = ((series < lower) | (series > upper)).sum()
    outlier_report.append((col, int(outliers), int(series.shape[0]), round(100*outliers/series.shape[0], 2)))

outlier_df = pd.DataFrame(outlier_report, columns=["column","outlier_count","n","outlier_%"]).sort_values("outlier_%", ascending=False)

print("✅ Aykırı değer özeti (en çoktan aza):")
display(outlier_df.head(15))
```

🧭 **Sonraki adım:**  
**Adım 4 — Kodlama (Encoding) ve Ölçekleme (Scaling)**  
Bu aşamada:
- `Attrition` sütununu 1/0’a dönüştüreceğiz,  
- Kategorik sütunları **One-Hot Encoding**,  
- Sayısal sütunları **StandardScaler** ile ölçeklendireceğiz.


```python
# 🔹 Adım 4 — Kodlama (Encoding) ve Ölçekleme (Scaling)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1️⃣ Kopya veriyle çalışalım
df_enc = df.copy()

# 2️⃣ Hedef değişkeni 1/0'a çevir (Yes/No)
df_enc["Attrition"] = df_enc["Attrition"].map({"Yes": 1, "No": 0}).astype("int64")

# 3️⃣ Çıkarılacak sütunlar (bilgi taşımayan)
drop_cols = ["EmpID", "EmployeeCount", "StandardHours", "Over18"]
df_enc = df_enc.drop(columns=drop_cols)

# 4️⃣ Giriş ve çıkış ayrımı
X = df_enc.drop(columns=["Attrition"])
y = df_enc["Attrition"]

# 5️⃣ Tür tespiti
num_cols_X = X.select_dtypes(include=["int64","float64"]).columns.tolist()
cat_cols_X = X.select_dtypes(include=["object","category"]).columns.tolist()

print("✅ Sayısal sütun örnekleri:", num_cols_X[:10])
print("✅ Kategorik sütun örnekleri:", cat_cols_X[:10])

# 6️⃣ Ön işleme pipeline
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols_X),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols_X)
    ],
    remainder="drop"
)

# 7️⃣ Eğitim/test ayrımı
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 8️⃣ Pipeline fit
pipeline_preview = Pipeline(steps=[("preprocess", preprocess)])
pipeline_preview.fit(X_train, y_train)

# 9️⃣ Dönüşmüş matris boyutu
X_train_t = pipeline_preview.transform(X_train)
X_test_t = pipeline_preview.transform(X_test)

print("✅ Dönüştürülmüş eğitim matris boyutu:", X_train_t.shape)
print("✅ Dönüştürülmüş test matris boyutu:", X_test_t.shape)
```

### ⚙️ Adım 4 — Kodlama (Encoding) ve Ölçekleme (Scaling)

**Amaç:**  
Modelleme öncesinde sayısal ve kategorik değişkenleri makine öğrenmesi algoritmalarının anlayacağı forma dönüştürmek.

---

#### 🔹 Yaptığımız İşlemler

1. **Hedef değişken (`Attrition`) dönüştürüldü:**  
   “Yes” → 1, “No” → 0 olarak sayısallaştırıldı.  

2. **Gereksiz sütunlar çıkarıldı:**  
   `EmpID`, `EmployeeCount`, `StandardHours`, `Over18`  
   (bunlar sabit veya kimlik sütunlarıydı).  

3. **Sayısal ve kategorik sütunlar ayrıldı:**
   - **Sayısal örnekler:** `Age`, `DailyRate`, `DistanceFromHome`, `Education`, `HourlyRate`, `JobSatisfaction`  
   - **Kategorik örnekler:** `BusinessTravel`, `Department`, `Gender`, `JobRole`, `OverTime`, `SalarySlab`  

4. **Ön işleme (Preprocessing) pipeline’ı oluşturuldu:**
   - Sayısallar → `StandardScaler()` ile **ölçeklendi** (ortalama=0, varyans=1)  
   - Kategorikler → `OneHotEncoder()` ile **dummy sütunlara dönüştürüldü**  
   - Pipeline yalnızca **eğitim verisi üzerinde fit edildi** (data leakage önlendi)

---

#### 📏 Dönüşüm Sonrası Veri Boyutu

| Veri Seti | Satır Sayısı | Özellik Sayısı |
|------------|--------------|----------------|
| Eğitim (`X_train_t`) | 1184 | **62** |
| Test (`X_test_t`) | 296 | **62** |

🔹 62 özellik, one-hot encoding sonrası sütun sayısının arttığını gösteriyor.  
Bu, kategorik sütunların her bir benzersiz değeri için yeni bir sütun oluştuğu anlamına gelir.  

---

#### ✅ Sonuç:
- Encoding ve ölçekleme işlemleri başarıyla tamamlandı.  
- Veri, modellemeye hazır hale geldi.  
- Sayısal değerler aynı ölçek düzeyinde, kategorikler binary formatta.

```python
# 🔹 Adım 5 — Görselleştirme ve Korelasyon Analizi

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", font_scale=1.0)

# 1️⃣ Attrition dağılımı
plt.figure(figsize=(5,4))
df["Attrition"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
plt.title("Attrition Dağılımı")
plt.xlabel("Sınıf")
plt.ylabel("Adet")
plt.show()

# 2️⃣ Örnek: birkaç kategorik değişkenin Attrition oranı
def stacked_ratio_plot(cat_col, target="Attrition"):
    ctab = pd.crosstab(df[cat_col], df[target], normalize="index") * 100
    ctab = ctab.reindex(ctab.sum(axis=1).sort_values(ascending=False).index)
    ctab.plot(kind="bar", stacked=True, figsize=(6,4), colormap="coolwarm")
    plt.title(f"{cat_col} Göre Attrition Oranı (%)")
    plt.ylabel("Yüzde")
    plt.legend(title=target, bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

for c in ["Department", "Gender", "OverTime"]:
    stacked_ratio_plot(c)

# 3️⃣ Sayısal dağılımlar (örnek 2 sütun)
for c in ["MonthlyIncome", "YearsAtCompany"]:
    fig, ax = plt.subplots(2,1, figsize=(6,6))
    sns.histplot(df[c].dropna(), ax=ax[0], kde=True)
    ax[0].set_title(f"{c} - Histogram")
    sns.boxplot(x=df[c].dropna(), ax=ax[1])
    ax[1].set_title(f"{c} - Boxplot")
    plt.tight_layout()
    plt.show()

# 4️⃣ Korelasyon (sayısallar arası)
num_for_corr = df.select_dtypes(include=["int64","float64"]).copy()
num_for_corr["Attrition"] = df["Attrition"].map({"Yes":1,"No":0}).astype("float64")

plt.figure(figsize=(10,8))
corr = num_for_corr.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Sayısallar Arası Korelasyon Isı Haritası")
plt.show()

# 5️⃣ Attrition ile en ilişkili sayısallar
target_corr = corr["Attrition"].drop("Attrition").sort_values(ascending=False)
print("✅ Attrition ile en güçlü pozitif korelasyonlar:")
display(target_corr.head(10))
print("✅ Attrition ile en güçlü negatif korelasyonlar:")
display(target_corr.tail(10))
```

```python
# === IQR kapama + etki raporu (sağlam kurulumlu) ===
import numpy as np
import pandas as pd

# 0) Ön koşullar: df ve hedef kolon var mı?
assert 'df' in globals(), "Önce veri yükleme hücresini çalıştırıp df'i oluşturmalısın."
target_col = "Attrition"
assert target_col in df.columns, f"{target_col} kolonu df içinde bulunamadı."

# 1) df_cleaned yoksa yeniden kur
if 'df_cleaned' not in globals():
    drop_cols = ['EmployeeCount', 'StandardHours', 'Over18', 'EmpID', 'EmployeeNumber']
    df_cleaned = df.drop(columns=drop_cols, errors='ignore').copy()
    if 'YearsWithCurrManager' in df_cleaned.columns:
        df_cleaned['YearsWithCurrManager'] = df_cleaned['YearsWithCurrManager'] \
            .fillna(df_cleaned['YearsWithCurrManager'].median())

print("✅ df_cleaned hazır. Şekil:", df_cleaned.shape)

# 2) IQR kapama (clip) ve etki raporu
df_capped = df_cleaned.copy()
num_cols = df_capped.select_dtypes(include=["int64","float64"]).columns.tolist()

iqr_rows = []
for col in num_cols:
    s = df_capped[col].dropna()
    if s.empty:
        continue
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    before = df_capped[col].copy()
    df_capped[col] = df_capped[col].clip(lower=lower, upper=upper)

    affected = int((before != df_capped[col]).sum())
    iqr_rows.append({
        "column": col,
        "Q1": float(q1), "Q3": float(q3), "IQR": float(iqr),
        "lower": float(lower), "upper": float(upper),
        "affected_count": affected,
        "affected_%": round(100 * affected / len(before.dropna()), 2) if len(before.dropna())>0 else 0.0
    })

iqr_df = pd.DataFrame(iqr_rows).sort_values("affected_count", ascending=False)

print("✅ IQR clip sonrası etki raporu (ilk 15):")
display(iqr_df.head(15))

# (İsteğe bağlı) Sonradan kullanmak için sakla:
# iqr_df.to_csv("reports/iqr_effect_report.csv", index=False)

# df_capped bundan sonra görselleştirmede kullanılabilir (boxplot vs.)
```

```python
# 📐 Cramér’s V: Kategorik değişkenlerin Attrition ile ilişki gücü
import numpy as np
import scipy.stats as ss

cat_cols = df_cleaned.select_dtypes(include=["object","category","bool"]).columns.tolist()
cat_cols = [c for c in cat_cols if c != "Attrition"]

def cramers_v(x, y):
    confusion = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion)[0]
    n = confusion.sum().sum()
    phi2 = chi2 / n
    r, k = confusion.shape
    return np.sqrt(phi2 / max(1, min(k-1, r-1)))

cv_scores = {col: cramers_v(df_cleaned[col], df_cleaned["Attrition"]) for col in cat_cols}
cv_series = pd.Series(cv_scores).sort_values(ascending=False)
display(cv_series.head(10))

# Kaydet
os.makedirs("reports", exist_ok=True)
cv_series.to_csv("reports/categorical_strength_cramersV.csv", header=["cramers_v"])
print("✅ Kaydedildi: reports/categorical_strength_cramersV.csv")
```
